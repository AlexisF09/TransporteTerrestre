```python
# ============================================
# Autopista Panamá–Colón
# ============================================

# 1. Librerías
import os 
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import triang
from datetime import datetime

# Configuración visual
plt.style.use("seaborn-v0_8")
sns.set(rc={"figure.figsize": (12, 6)})
pd.set_option("display.max_columns", None)

# ================================
# 2. Cargar los datos
# ================================
# Ajusta el nombre del archivo a tu dataset
downloads_folder = os.path.join(os.path.expanduser("~"), "Downloads")
file = os.path.join(downloads_folder, "rutas_PanamaColon_con_clima_20250914.csv")
df = pd.read_csv(file)

# Vista inicial
print("Dimensiones:", df.shape)
display(df.head(3))
print(df.info())

# ================================
# 3. Análisis descriptivo inicial
# ================================
print("\n--- Estadística descriptiva de duraciones ---")
display(df[["duration_min(minutes)", "duration_max(minutes)", "duration(minutes)"]].describe())

print("\n--- Estadística descriptiva de clima en origen ---")
display(df[["origin_temperature_2m", "origin_precipitation", "origin_windspeed_10m"]].describe())

# Revisar consistencia: promedio vs (min+max)/2
df["duration_mid"] = (df["duration_min(minutes)"] + df["duration_max(minutes)"]) / 2
df["diff_avg_mid"] = df["duration(minutes)"] - df["duration_mid"]
print("\nDiferencia promedio vs midpoint:", df["diff_avg_mid"].mean())

# ================================
# 4. Distribuciones básicas
# ================================
fig, axes = plt.subplots(1, 3, figsize=(18, 5))
sns.histplot(df["duration_min(minutes)"], bins=20, ax=axes[0], color="skyblue")
axes[0].set_title("Duración mínima")

sns.histplot(df["duration_max(minutes)"], bins=20, ax=axes[1], color="orange")
axes[1].set_title("Duración máxima")

sns.histplot(df["duration(minutes)"], bins=20, ax=axes[2], color="green")
axes[2].set_title("Duración promedio")
plt.show()

# Boxplot
sns.boxplot(data=df[["duration_min(minutes)", "duration(minutes)", "duration_max(minutes)"]])
plt.title("Boxplot de Duraciones")
plt.show()

# ================================
# 5. Análisis temporal
# ================================
df["datetime_utc"] = pd.to_datetime(df["datetime_utc"])
df["hour"] = df["datetime_utc"].dt.hour
df["dayofweek"] = df["datetime_utc"].dt.dayofweek  # 0=Lunes, 6=Domingo

# Promedio de duración por hora
avg_by_hour = df.groupby("hour")["duration(minutes)"].mean()
avg_by_hour.plot(marker="o", title="Duración promedio por hora del día")
plt.xlabel("Hora del día")
plt.ylabel("Duración promedio (min)")
plt.show()

# ================================
# 6. Relación clima y duración
# ================================
sns.scatterplot(data=df, x="origin_precipitation", y="duration(minutes)", alpha=0.6)
plt.title("Duración vs Precipitación")
plt.show()

sns.scatterplot(data=df, x="origin_windspeed_10m", y="duration(minutes)", alpha=0.6)
plt.title("Duración vs Viento (10m)")
plt.show()

# Heatmap de correlaciones
corr = df[[
    "duration(minutes)", "duration_min(minutes)", "duration_max(minutes)",
    "origin_temperature_2m", "origin_precipitation", "origin_windspeed_10m",
    "destination_temperature_2m", "destination_precipitation", "destination_windspeed_10m"
]].corr()

sns.heatmap(corr, annot=True, cmap="coolwarm", center=0)
plt.title("Matriz de correlaciones")
plt.show()

# ================================
# 7. Distribución triangular (simulación)
# ================================
# Ejemplo con la primera fila
row = df.iloc[0]
c = (row["duration(minutes)"] - row["duration_min(minutes)"]) / (row["duration_max(minutes)"] - row["duration_min(minutes)"])
r = triang(c, loc=row["duration_min(minutes)"], scale=row["duration_max(minutes)"] - row["duration_min(minutes)"])

simulated = r.rvs(1000)  # 1000 muestras simuladas

sns.histplot(simulated, bins=30, kde=True, color="purple")
plt.axvline(row["duration_min(minutes)"], color="blue", linestyle="--", label="Min")
plt.axvline(row["duration(minutes)"], color="green", linestyle="--", label="Promedio")
plt.axvline(row["duration_max(minutes)"], color="red", linestyle="--", label="Max")
plt.title("Distribución Triangular simulada (ejemplo)")
plt.legend()
plt.show()

# ================================
# 8. Features derivados
# ================================
df["speed_range"] = df["duration_max(minutes)"] - df["duration_min(minutes)"]
df["congestion_index"] = df["duration(minutes)"] / df["duration_max(minutes)"]

sns.scatterplot(data=df, x="congestion_index", y="duration(minutes)", alpha=0.6)
plt.title("Duración vs Índice de Congestión")
plt.show()

print("EDA completado 🚀")
```
